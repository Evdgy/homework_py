```python
# Пре-процессинг
import re 
with open ('tiny-shakespeare.txt', 'r') as f:
    data = f.read()

with open ('stopwords.txt','r') as f:
    sw = f.read()
```

```python
print(data[:100])
```

```python
print(sw[:100])
```

```python
# Удвлите пунктцацию
def remove_punctuation(text):
    pattern = r'[^\w\s]'
    cleaned_text = re.sub(pattern,'', text)
    return cleaned_text
remove_punctuation('This is a sample text! Wow :-)')
```

```python
# Удалить стоп-слова

def remove_stopwords(text, stopwords):
    
    stopwords = sw.split()
    words = text.split()
    filtered_words = [] # создайте пустой список
    for word in words:
        if word.lower() not in stopwords:# если слово, приведенное к нижнему регистру присутствует в списке стоп-слов,
            filtered_words.append(word) # объедините слова из списка filtered_words в строку с join
    cleaned_text = ' '.join(filtered_words)
    return cleaned_text

remove_stopwords('Below is the text without stopwords', sw)
```

```python
# Токенизировать текст

def tokenize_text(text):
    text = text.lower()
    tokenized_text = text.split('\n')# разделить по переносу строки
    return tokenized_text

# Протестировать функцию на тестовом образце
tokenize_text('Tokenize this text\nWe\'ve got to tokenize it')
```

```python
# Применить функцию токенизации к данным
tokenized_data = tokenize_text(data)

print(f'{type(tokenized_data)}\n') #Вывести на экран тип объекта tokenized_data
print(f'{len(tokenized_data)}\n') #Вывести на экран количество элементов tokenized_data

print(f'{tokenized_data[0][:100]}\n')# Вывести на экран первые 100 элементов первого объекта tokenized_data

print(f'{tokenized_data[1][:100]}\n')# Вывести на экран первые 100 элементов второго объекта tokenized_data
```

```python
cleaned_data = []
for element in tokenized_data:
    cleaned_data.append(remove_stopwords(element, sw))# примените функцию для очистки текстов от стоп-слов к каждому элементу списка cleaned_data
# не забудьте передать в функцию стоп-слова
print(f'{cleaned_data[0][:100]}')
print(f'{cleaned_data[1][:100]}')
```

```python
# Подготовка данных для генеративной модели
formatted_texts = []
for text in cleaned_data:
    if len(text) > 0:
        formatted_text = f'<SoS>{text.strip()}<EoS>'
        formatted_texts.append(formatted_text)
        
formatted_texts[:100]     
```

```python
# Сохраните результат в новый *.txt
with open('shakespeare_gpt2', 'w') as f:
    f.write('\n'.join(formatted_texts)) # режим для записи, используйте join и разделитель \n
```

## Лингвистический анализ 

```python
# Небольшая выборка для аннотирования данных
import random 
filtered_data = []
for text in cleaned_data:
    if len(text) > 0:
        filtered_data.append(text)
        
sample_size = max(1,int(len(filtered_data) * 0.1))  # задаем размер выборки 10% от общего количества элементов 

sample_data = random.sample(filtered_data, sample_size) # семплирование

sample_text = ''.join(str(sample_data))
sample_text
```

```python
import spacy

nlp = spacy.load("en_core_web_sm") # подгружаем модель

doc = nlp(sample_text)

lemmas = []

for token in doc:
  lemmas.append(token.lemma_)

for i in range(30):
  print(doc[i], lemmas[i])
# выведите пары (token, lemma) в диапазоне 30,
# т.е. для первых 30 элементов списка lemmas и для первых 30 элементов doc
 
```

```python
# Частеречная разметка
pos_annotation = []

for token in doc:
    pos_annotation.append((token.text, token.pos_))
for i in range(30):
    print(f'{pos_annotation[i][0]} -> {pos_annotation[i][1]}') # выведите пары (token, pos) в диапазоне 30
```

```python
# Синтаксическая разметка
from spacy import displacy

doc_ = nlp('This is a sample sentence.')

syntax_annotation = [(token.text, token.dep_, token.head.text) for token in doc_]
displacy.render(doc_, style="dep", jupyter=True, options={'distance': 150})
```

```python
# Разметка Named Entity Recognition
ner_annotation = [(ent.text, ent.label_) for ent in doc.ents]
displacy.render(doc, style="ent", jupyter=True)
```

```python
# Создайте словарь с результатами разметки
from pprint import pprint
results = dict()
results['lemmas'] = lemmas[:10]
results['pos'] = pos_annotation[:10]

pprint(results)
```
